```python
import torch 
from torch import nn
from d2l import torch as d2l
```

**ç”Ÿæˆæ•°æ®é›†**

* ç®€å•èµ·è§ï¼Œè€ƒè™‘ä¸‹é¢è¿™ä¸ªå›å½’é—®é¢˜ï¼šç»™å®šçš„æˆå¯¹çš„â€œè¾“å…¥ï¼è¾“å‡ºâ€æ•°æ®é›†$\{(x_1, y_1), \ldots, (x_n, y_n)\}$ï¼Œå¦‚ä½•å­¦ä¹ $f$æ¥é¢„æµ‹ä»»æ„æ–°è¾“å…¥$x$çš„è¾“å‡º$\hat{y} = f(x)$ï¼Ÿ

* æ ¹æ®ä¸‹é¢çš„éçº¿æ€§å‡½æ•°ç”Ÿæˆä¸€ä¸ªäººå·¥æ•°æ®é›†ï¼Œå…¶ä¸­åŠ å…¥çš„å™ªå£°é¡¹ä¸º$\epsilon$ï¼š

   $$y_i = 2\sin(x_i) + x_i^{0.8} + \epsilon,$$

* å…¶ä¸­$\epsilon$æœä»å‡å€¼ä¸º$0$å’Œæ ‡å‡†å·®ä¸º$0.5$çš„æ­£æ€åˆ†å¸ƒã€‚ç”Ÿæˆäº†$50$ä¸ªè®­ç»ƒæ ·æœ¬å’Œ$50$ä¸ªæµ‹è¯•æ ·æœ¬ã€‚ä¸ºäº†æ›´å¥½åœ°å¯è§†åŒ–ä¹‹åçš„æ³¨æ„åŠ›æ¨¡å¼ï¼Œå°†è®­ç»ƒæ ·æœ¬è¿›è¡Œæ’åºã€‚

```python
n_train = 50  # è®­ç»ƒæ ·æœ¬æ•°
x_train, _ = torch.sort(torch.rand(n_train) * 5)   # æ’åºåçš„è®­ç»ƒæ ·æœ¬

def f(x):
    return 2 * torch.sin(x) + x**0.8

y_train = f(x_train) + torch.normal(0.0, 0.5, (n_train,))  # è®­ç»ƒæ ·æœ¬çš„è¾“å‡º
x_test = torch.arange(0, 5, 0.1)  # æµ‹è¯•æ ·æœ¬
y_truth = f(x_test)  # æµ‹è¯•æ ·æœ¬çš„çœŸå®è¾“å‡º
n_test = len(x_test)  # æµ‹è¯•æ ·æœ¬æ•°
n_test
```

* ä¸‹é¢çš„å‡½æ•°å°†ç»˜åˆ¶æ‰€æœ‰çš„è®­ç»ƒæ ·æœ¬ï¼ˆæ ·æœ¬ç”±åœ†åœˆè¡¨ç¤ºï¼‰ï¼Œä¸å¸¦å™ªå£°é¡¹çš„çœŸå®æ•°æ®ç”Ÿæˆå‡½æ•°$f$ï¼ˆæ ‡è®°ä¸ºâ€œTruthâ€ï¼‰ï¼Œä»¥åŠå­¦ä¹ å¾—åˆ°çš„é¢„æµ‹å‡½æ•°ï¼ˆæ ‡è®°ä¸ºâ€œPredâ€ï¼‰ã€‚

```python
def plot_kernel_reg(y_hat):
    d2l.plot(x_test, [y_truth, y_hat], 'x', 'y', legend=['Truth', 'Pred'],
             xlim=[0, 5], ylim=[-1, 5])
    d2l.plt.plot(x_train, y_train, 'o', alpha=0.5);
```

**å¹³å‡æ± åŒ–**

* å…ˆä½¿ç”¨æœ€ç®€å•çš„ä¼°è®¡å™¨æ¥è§£å†³å›å½’é—®é¢˜ï¼šåŸºäºå¹³å‡æ±‡èšæ¥è®¡ç®—æ‰€æœ‰è®­ç»ƒæ ·æœ¬è¾“å‡ºå€¼çš„å¹³å‡å€¼ï¼š

$$f(x) = \frac{1}{n}\sum_{i=1}^n y_i$$

* å¦‚ä¸‹å›¾æ‰€ç¤ºï¼Œè¿™ä¸ªä¼°è®¡å™¨ç¡®å®ä¸å¤Ÿèªæ˜ï¼š çœŸå®å‡½æ•° ğ‘“ ï¼ˆâ€œTruthâ€ï¼‰å’Œé¢„æµ‹å‡½æ•°ï¼ˆâ€œPredâ€ï¼‰ç›¸å·®å¾ˆå¤§ã€‚

```python
y_hat = torch.repeat_interleave(y_train.mean(), n_test)
plot_kernel_reg(y_hat)
```

**éå‚æ•°æ³¨æ„åŠ›æ± åŒ–**

* æ˜¾ç„¶ï¼Œå¹³å‡æ±‡èšå¿½ç•¥äº†è¾“å…¥$x_i$ã€‚äºæ˜¯Nadaraya :cite:`Nadaraya.1964`å’Œ
Watson :cite:`Watson.1964`æå‡ºäº†ä¸€ä¸ªæ›´å¥½çš„æƒ³æ³•ï¼Œæ ¹æ®è¾“å…¥çš„ä½ç½®å¯¹è¾“å‡º$y_i$è¿›è¡ŒåŠ æƒï¼š

$$f(x) = \sum_{i=1}^n \frac{K(x - x_i)}{\sum_{j=1}^n K(x - x_j)} y_i$$

* å…¶ä¸­$K$æ˜¯**æ ¸**ï¼ˆkernelï¼‰ã€‚æ‰€æè¿°çš„ä¼°è®¡å™¨è¢«ç§°ä¸ºNadaraya-Watsonæ ¸å›å½’ï¼ˆNadaraya-Watson kernel regressionï¼‰ã€‚å—æ­¤å¯å‘ï¼Œå¯ä»¥ä»æ³¨æ„åŠ›æœºåˆ¶æ¡†æ¶çš„è§’åº¦é‡å†™`Nadaraya-Watsonæ ¸å‡½æ•°`ï¼Œæˆä¸ºä¸€ä¸ªæ›´åŠ é€šç”¨çš„æ³¨æ„åŠ›æ± åŒ–ï¼ˆattention poolingï¼‰å…¬å¼ï¼š

$$f(x) = \sum_{i=1}^n \alpha(x, x_i) y_i,$$


* **å…¶ä¸­$x$æ˜¯æŸ¥è¯¢ï¼Œ$(x_i, y_i)$æ˜¯é”®å€¼å¯¹ã€‚æ³¨æ„åŠ›æ±‡èšæ˜¯$y_i$çš„åŠ æƒå¹³å‡ã€‚**
å°†æŸ¥è¯¢$x$å’Œé”®$x_i$ä¹‹é—´çš„å…³ç³»å»ºæ¨¡ä¸º**æ³¨æ„åŠ›æƒé‡**ï¼ˆattention weightï¼‰$\alpha(x, x_i)$ï¼Œ
è¿™ä¸ªæƒé‡å°†è¢«åˆ†é…ç»™æ¯ä¸€ä¸ªå¯¹åº”å€¼$y_i$ã€‚å¯¹äºä»»ä½•æŸ¥è¯¢ï¼Œæ¨¡å‹åœ¨æ‰€æœ‰é”®å€¼å¯¹æ³¨æ„åŠ›æƒé‡éƒ½æ˜¯ä¸€ä¸ªæœ‰æ•ˆçš„æ¦‚ç‡åˆ†å¸ƒï¼šå®ƒä»¬æ˜¯éè´Ÿçš„ï¼Œå¹¶ä¸”æ€»å’Œä¸º1ã€‚

* ä¸ºäº†æ›´å¥½åœ°ç†è§£æ³¨æ„åŠ›æ±‡èšï¼Œè€ƒè™‘ä¸€ä¸ª**é«˜æ–¯æ ¸**ï¼ˆGaussian kernelï¼‰ï¼Œå…¶å®šä¹‰ä¸ºï¼š$$K(u) = \frac{1}{\sqrt{2\pi}} \exp(-\frac{u^2}{2})$$å°†é«˜æ–¯æ ¸ä»£å…¥å¯ä»¥å¾—åˆ°ï¼š
$$\begin{aligned} f(x) &=\sum_{i=1}^n \alpha(x, x_i) y_i\\ &= \sum_{i=1}^n \frac{\exp\left(-\frac{1}{2}(x - x_i)^2\right)}{\sum_{j=1}^n \exp\left(-\frac{1}{2}(x - x_j)^2\right)} y_i \\&= \sum_{i=1}^n \mathrm{softmax}\left(-\frac{1}{2}(x - x_i)^2\right) y_i. \end{aligned}$$

* å¦‚æœä¸€ä¸ªé”®$x_i$è¶Šæ˜¯æ¥è¿‘ç»™å®šçš„æŸ¥è¯¢$x$ï¼Œé‚£ä¹ˆåˆ†é…ç»™è¿™ä¸ªé”®å¯¹åº”å€¼$y_i$çš„æ³¨æ„åŠ›æƒé‡å°±ä¼šè¶Šå¤§ï¼Œ
ä¹Ÿå°±â€œè·å¾—äº†æ›´å¤šçš„æ³¨æ„åŠ›â€ã€‚å€¼å¾—æ³¨æ„çš„æ˜¯ï¼ŒNadaraya-Watsonæ ¸å›å½’æ˜¯ä¸€ä¸ªéå‚æ•°æ¨¡å‹ã€‚
å› æ­¤ï¼Œ`Nadaraya-Watson-Gaussian`æ˜¯**éå‚æ•°çš„æ³¨æ„åŠ›æ±‡æ± åŒ–**ï¼ˆnonparametric attention poolingï¼‰æ¨¡å‹ã€‚

* æ¥ä¸‹æ¥ï¼Œå°†åŸºäºè¿™ä¸ªéå‚æ•°çš„æ³¨æ„åŠ›æ± åŒ–æ¨¡å‹æ¥ç»˜åˆ¶é¢„æµ‹ç»“æœã€‚æ–°çš„æ¨¡å‹é¢„æµ‹çº¿æ˜¯å¹³æ»‘çš„ï¼Œå¹¶ä¸”æ¯”å¹³å‡æ± åŒ–çš„é¢„æµ‹æ›´æ¥è¿‘çœŸå®ã€‚

```python
# X_repeatçš„å½¢çŠ¶:(n_test,n_train),
# æ¯ä¸€è¡Œéƒ½åŒ…å«ç€ç›¸åŒçš„æµ‹è¯•è¾“å…¥ï¼ˆä¾‹å¦‚ï¼šåŒæ ·çš„æŸ¥è¯¢ï¼‰
X_repeat = x_test.repeat_interleave(n_train).reshape((-1, n_train))
# x_trainåŒ…å«ç€é”®ã€‚attention_weightsçš„å½¢çŠ¶ï¼š(n_test,n_train),
# æ¯ä¸€è¡Œéƒ½åŒ…å«ç€è¦åœ¨ç»™å®šçš„æ¯ä¸ªæŸ¥è¯¢çš„å€¼ï¼ˆy_trainï¼‰ä¹‹é—´åˆ†é…çš„æ³¨æ„åŠ›æƒé‡
attention_weights = nn.functional.softmax(-(X_repeat - x_train)**2 / 2, dim=1)
# y_hatçš„æ¯ä¸ªå…ƒç´ éƒ½æ˜¯å€¼çš„åŠ æƒå¹³å‡å€¼ï¼Œå…¶ä¸­çš„æƒé‡æ˜¯æ³¨æ„åŠ›æƒé‡
y_hat = torch.matmul(attention_weights, y_train)
plot_kernel_reg(y_hat)
```

**æ³¨æ„åŠ›æƒé‡**

* æµ‹è¯•æ•°æ®çš„è¾“å…¥ç›¸å½“äºæŸ¥è¯¢ï¼Œè€Œè®­ç»ƒæ•°æ®çš„è¾“å…¥ç›¸å½“äºé”®ã€‚å› ä¸ºä¸¤ä¸ªè¾“å…¥éƒ½æ˜¯ç»è¿‡æ’åºçš„ï¼Œå› æ­¤ç”±è§‚å¯Ÿå¯çŸ¥â€œæŸ¥è¯¢-é”®â€å¯¹è¶Šæ¥è¿‘ï¼Œæ³¨æ„åŠ›æ±‡èšçš„**æ³¨æ„åŠ›æƒé‡**å°±è¶Šé«˜ã€‚

```python
d2l.show_heatmaps(attention_weights.unsqueeze(0).unsqueeze(0),
                  xlabel='Sorted training inputs',
                  ylabel='Sorted testing inputs')
```

**å¸¦å‚æ•°æ³¨æ„åŠ›æ±‡èš**

* éå‚æ•°çš„Nadaraya-Watsonæ ¸å›å½’å…·æœ‰*ä¸€è‡´æ€§*ï¼ˆconsistencyï¼‰çš„ä¼˜ç‚¹ï¼š
å¦‚æœæœ‰è¶³å¤Ÿçš„æ•°æ®ï¼Œæ­¤æ¨¡å‹ä¼šæ”¶æ•›åˆ°æœ€ä¼˜ç»“æœã€‚å°½ç®¡å¦‚æ­¤ï¼Œè¿˜æ˜¯å¯ä»¥è½»æ¾åœ°å°†å¯å­¦ä¹ çš„å‚æ•°é›†æˆåˆ°æ³¨æ„åŠ›æ±‡èšä¸­ã€‚ä¾‹å¦‚ï¼Œåœ¨ä¸‹é¢çš„æŸ¥è¯¢$x$å’Œé”®$x_i$ä¹‹é—´çš„è·ç¦»ä¹˜ä»¥å¯å­¦ä¹ å‚æ•°$w$ï¼š

$$\begin{aligned}f(x) &= \sum_{i=1}^n \alpha(x, x_i) y_i \\&= \sum_{i=1}^n \frac{\exp\left(-\frac{1}{2}((x - x_i)w)^2\right)}{\sum_{j=1}^n \exp\left(-\frac{1}{2}((x - x_j)w)^2\right)} y_i \\&= \sum_{i=1}^n \mathrm{softmax}\left(-\frac{1}{2}((x - x_i)w)^2\right) y_i.\end{aligned}$$

**æ‰¹é‡çŸ©é˜µä¹˜æ³•**

* ä¸ºäº†æ›´æœ‰æ•ˆåœ°è®¡ç®—å°æ‰¹é‡æ•°æ®çš„æ³¨æ„åŠ›ï¼Œå¯ä»¥åˆ©ç”¨æ·±åº¦å­¦ä¹ å¼€å‘æ¡†æ¶ä¸­æä¾›çš„æ‰¹é‡çŸ©é˜µä¹˜æ³•ã€‚

* å‡è®¾ç¬¬ä¸€ä¸ªå°æ‰¹é‡æ•°æ®åŒ…å«$n$ä¸ªçŸ©é˜µ$\mathbf{X}_1,\ldots, \mathbf{X}_n$ï¼Œ
å½¢çŠ¶ä¸º$a\times b$ç¬¬äºŒä¸ªå°æ‰¹é‡åŒ…å«$n$ä¸ªçŸ©é˜µ$\mathbf{Y}_1, \ldots, \mathbf{Y}_n$ï¼Œ
å½¢çŠ¶ä¸º$b\times c$ã€‚å®ƒä»¬çš„æ‰¹é‡çŸ©é˜µä¹˜æ³•å¾—åˆ°$n$ä¸ªçŸ©é˜µ
$\mathbf{X}_1\mathbf{Y}_1, \ldots, \mathbf{X}_n\mathbf{Y}_n$ï¼Œ
å½¢çŠ¶ä¸º$a\times c$ã€‚å› æ­¤ï¼Œ**å‡å®šä¸¤ä¸ªå¼ é‡çš„å½¢çŠ¶åˆ†åˆ«æ˜¯$(n,a,b)$å’Œ$(n,b,c)$ï¼Œ
å®ƒä»¬çš„æ‰¹é‡çŸ©é˜µä¹˜æ³•è¾“å‡ºçš„å½¢çŠ¶ä¸º$(n,a,c)$**ã€‚

```python
X = torch.ones((2, 1, 4))
Y = torch.ones((2, 4, 6))
torch.bmm(X, Y).shape
```

* **ä½¿ç”¨å°æ‰¹é‡çŸ©é˜µä¹˜æ³•æ¥è®¡ç®—å°æ‰¹é‡æ•°æ®ä¸­çš„åŠ æƒå¹³å‡å€¼ã€‚**

```python
weights = torch.ones((2, 10)) * 0.1
values = torch.arange(20.0).reshape((2, 10))
torch.bmm(weights.unsqueeze(1), values.unsqueeze(-1))
```

**å®šä¹‰å¸¦å‚æ•°çš„æ³¨æ„åŠ›æ± åŒ–æ¨¡å‹**

```python
class NWKernelRegression(nn.Module):
    def __init__(self, **kwargs):
        super().__init__(**kwargs)
        self.w = nn.Parameter(torch.rand((1,), requires_grad=True))

    def forward(self, queries, keys, values):
        # querieså’Œattention_weightsçš„å½¢çŠ¶ä¸º(æŸ¥è¯¢ä¸ªæ•°ï¼Œâ€œé”®ï¼å€¼â€å¯¹ä¸ªæ•°)
        queries = queries.repeat_interleave(keys.shape[1]).reshape((-1, keys.shape[1]))
        self.attention_weights = nn.functional.softmax(
            -((queries - keys) * self.w)**2 / 2, dim=1)
        # valuesçš„å½¢çŠ¶ä¸º(æŸ¥è¯¢ä¸ªæ•°ï¼Œâ€œé”®ï¼å€¼â€å¯¹ä¸ªæ•°)
        return torch.bmm(self.attention_weights.unsqueeze(1),
                         values.unsqueeze(-1)).reshape(-1)
```

**è®­ç»ƒæ¨¡å‹**

* è®­ç»ƒå¸¦å‚æ•°çš„æ³¨æ„åŠ›æ±‡èšæ¨¡å‹æ—¶ï¼Œä½¿ç”¨å¹³æ–¹æŸå¤±å‡½æ•°å’Œéšæœºæ¢¯åº¦ä¸‹é™ã€‚

```python
# X_tileçš„å½¢çŠ¶:(n_trainï¼Œn_train)ï¼Œæ¯ä¸€è¡Œéƒ½åŒ…å«ç€ç›¸åŒçš„è®­ç»ƒè¾“å…¥
X_tile = x_train.repeat((n_train, 1))
# Y_tileçš„å½¢çŠ¶:(n_trainï¼Œn_train)ï¼Œæ¯ä¸€è¡Œéƒ½åŒ…å«ç€ç›¸åŒçš„è®­ç»ƒè¾“å‡º
Y_tile = y_train.repeat((n_train, 1))
# keysçš„å½¢çŠ¶:('n_train'ï¼Œ'n_train'-1)
keys = X_tile[(1 - torch.eye(n_train)).type(torch.bool)].reshape((n_train, -1))
# valuesçš„å½¢çŠ¶:('n_train'ï¼Œ'n_train'-1)
values = Y_tile[(1 - torch.eye(n_train)).type(torch.bool)].reshape((n_train, -1))
```

```python
net = NWKernelRegression()
loss = nn.MSELoss(reduction='none')
trainer = torch.optim.SGD(net.parameters(), lr=0.5)
animator = d2l.Animator(xlabel='epoch', ylabel='loss', xlim=[1, 5])

for epoch in range(5):
    trainer.zero_grad()
    l = loss(net(x_train, keys, values), y_train)
    l.sum().backward()
    trainer.step()
    print(f'epoch {epoch + 1}, loss {float(l.sum()):.6f}')
    animator.add(epoch + 1, float(l.sum()))
```

**é¢„æµ‹ç»“æœç»˜åˆ¶**

```python
# keysçš„å½¢çŠ¶:(n_testï¼Œn_train)ï¼Œæ¯ä¸€è¡ŒåŒ…å«ç€ç›¸åŒçš„è®­ç»ƒè¾“å…¥ï¼ˆä¾‹å¦‚ï¼Œç›¸åŒçš„é”®ï¼‰
keys = x_train.repeat((n_test, 1))
# valueçš„å½¢çŠ¶:(n_testï¼Œn_train)
values = y_train.repeat((n_test, 1))
y_hat = net(x_test, keys, values).unsqueeze(1).detach()
plot_kernel_reg(y_hat)
```
